```python
bins = np.linspace(data["age"].min(), data["age"].max(), 6)
labels = [
    "Young Adults (20-29)",
    "Early Career (30-39)",
    "Mid Career (40-49)",
    "Senior Professionals (50-59)",
    "Late Career / Retired (60-70)"
]
```

```python
data["age_group"] = pd.cut(data["age"], bins=bins, labels=labels, include_lowest=True)
```

```python
data
```

```python
data['age_group'].value_counts()
```

```python
plt.hist(data['age'])
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
data.columns
```

```python
data.groupby("age_group", observed=False)["annual Salary"].mean().sort_values(ascending=True).plot(kind="bar")
plt.title("Annual Salary by Age group")
plt.xlabel("Age Group")
plt.ylabel("Annual Salary")
plt.show()
```

```python
data.columns
```

```python
data.groupby("age_group", observed=False)["car purchase amount"].max().sort_values(ascending=False).plot(kind="bar")
plt.title("Maximum amount sale for age group")
plt.xlabel("Age Group")
plt.ylabel("Car purchase amount")
plt.show()
```
